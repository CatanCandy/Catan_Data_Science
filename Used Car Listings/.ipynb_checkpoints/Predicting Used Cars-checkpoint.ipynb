DATA PREPROCESSING

```python
# No electric engines
df.drop(['exterior_color','interior_color','engine'],axis=1)
df.dropna(subset=['transmission','engine_size'],inplace=True)

median_list = ['min_mpg','max_mpg','first_owner','damaged','personal_using']
for item in median_list:
    med = df[item].median()
    df[item].fillna(med, inplace = True)

df = df.astype({'first_owner':'int64',
               'damaged':'int64',
               'personal_using':'int64'})

df['price'] = df['price'].replace(['ot Priced'],0)
df['price'] = df['price'].replace([0],df['price'].median())

mode_list = ["damaged","first_owner","personal_using","turbo",'alloy_wheels',"adaptive_cruise_control","navigation_system",
             "power_liftgate","backup_camera","keyless_start","remote_start","sunroof/moonroof","automatic_emergency_braking",
             "stability_control","leather_seats","memory_seat","third_row_seating","apple_car_play/android_auto","bluetooth",
             "usb_port","heated_seats","automatic_transmission"]
for item in mode_list:
    mode = df[item].mode()
    df[item].fillna(mode, inplace=True)

df["price"] = df["price"].astype("float")

df.info()
```
